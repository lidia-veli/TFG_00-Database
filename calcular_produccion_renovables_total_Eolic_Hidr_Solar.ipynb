```python
import pandas as pd
import numpy as np
```

## Importar datos

```python
# DATOS PRODUCCIÓN ENERGÉTICA RENOVABLE
df_solar = pd.read_csv('Produccion_Solar/data_Prod_Solar_TT_EU27_1980_2024.csv', index_col=0, parse_dates=True)
df_hidr = pd.read_csv('Produccion_Hidraulica/data_Prod_Hidr_TT_EU27_1980_2024.csv', index_col=0, parse_dates=True)
df_eolic = pd.read_csv('Produccion_Eolica/data_Prod_Eolic_TT_EU27_1980_2024.csv', index_col=0, parse_dates=True)
```

```python
# comprobar que las columnas tienen los mismo países
print(df_solar.columns.tolist() == df_hidr.columns.tolist() == df_eolic.columns.tolist())

# comprobar que los índices son iguales
print(df_solar.index.tolist() == df_hidr.index.tolist() == df_eolic.index.tolist())
```

```python
# rellenar NaN con 0.0p orque si no, la suma va a dar problemas
df_solar.fillna(0.0, inplace=True)
df_hidr.fillna(0.0, inplace=True)
df_eolic.fillna(0.0, inplace=True)
```

## Suma renovables

```python
# vamos a dejar sólo los países en la lista de PAISES_UE27
df_oferta_EU27 = pd.DataFrame(index=df_solar.index)

for pais in df_solar.columns:
    df_oferta_EU27[pais] = df_solar[pais] + df_hidr[pais] + df_eolic[pais]

df_oferta_EU27.tail()
```

```python
df_oferta_EU27.to_csv('data_Oferta_Renovables_EU27_1980_2024.csv', index=True)
```
