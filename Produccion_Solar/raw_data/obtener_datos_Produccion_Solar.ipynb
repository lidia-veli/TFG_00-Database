```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_PSF = pd.read_csv('raw_data_Produccion_Solar_Fotovoltaica_ENRGY.csv',
                         skiprows=52)
data_PSF.head()
```

## 1) PREPROCESAMIENTO DATOS

### Indexar por fecha

```python
# indexar por fecha
data_PSF.index = pd.to_datetime(data_PSF['Date'])
data_PSF.head()
```

```python
# quitar columna Date
data_PSF.drop(columns=['Date'], inplace=True)
```

```python
# lista de países en los datos
paises_PSF = data_PSF.columns.to_list()
print(len(paises_PSF))
print(paises_PSF)
```

### Datos nulos

```python
# crear un datatframe del número de valores nulos que tiene cada país
nulos_PSF = pd.DataFrame(data_PSF.isnull().sum(), columns=['Num_Null'])
#nulos_PSF

# NO HAY NULOS
```

```python
# veamos si hay datos con valor cero
nulos_PSF['Num_Zero'] = data_PSF[data_PSF == 0].count()

nulos_PSF_pais1 = nulos_PSF[nulos_PSF['Num_Zero'] == 0]
nulos_PSF_pais2 = nulos_PSF[nulos_PSF['Num_Zero'] > 0]
nulos_PSF_pais2  # con valores cero
```

```python
# comprobamos que la suma total de las filas _pais1 y _pais2 es igual al de nulos_PSF
bool_PSF = nulos_PSF['Num_Null'].sum() == nulos_PSF_pais1['Num_Null'].sum() + nulos_PSF_pais2['Num_Null'].sum()
print(bool_PSF)

print("Paises de interés")
print(nulos_PSF_pais1.index.tolist())
print()
print("Paises con todo datos vacíos")
print(nulos_PSF_pais2.index.tolist())
```
