```python
import pandas as pd
import numpy as np
```

## Importar datos

```python
# DATOS PRODUCCIÓN SOLAR FOTOVOLTAICA
df_PSF = pd.read_csv('data_PSF_1980_2024.csv', index_col=0, parse_dates=True)
                                    # tomamos la primer columna del csv como índice  en formato fecha
```

## Cobertura conjunta Europa

```python
PAISES_EU27 = [
    'AT', 'BE', 'BG', 'CY', 'CZ', 'DE', 'DK', 'EE', 'EL', 'ES', 'FI', 'FR', 
    'HR', 'HU', 'IE', 'IT', 'LT', 'LU', 'LV', 'MT', 'NL', 'PL', 'PT', 'RO', 
    'SE', 'SI', 'SK']
```

```python
# vamos a dejar sólo los países en la lista de PAISES_UE27
df_prod_solar_EU27 = df_PSF.copy()

for pais in df_prod_solar_EU27.columns:
    if pais not in PAISES_EU27:
        df_prod_solar_EU27.drop(pais, axis=1, inplace=True)
    else:
        continue

# y creamos la columna de la producción solar conjunta de los países de la UE27
df_prod_solar_EU27['EU27'] = df_prod_solar_EU27.sum(axis=1)

df_prod_solar_EU27.tail()
```

```python
df_prod_solar_EU27.to_csv('data_Prod_Solar_TT_EU27_1980_2024.csv', index=True)
```
