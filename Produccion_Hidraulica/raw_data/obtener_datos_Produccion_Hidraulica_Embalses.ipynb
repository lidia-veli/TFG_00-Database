```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_HRES = pd.read_csv('Datos_Produccion_Hidraulica_Embalses.csv',
                         skiprows=61)
data_HRES.head()
```

## 1) PREPROCESAMIENTO DATOS

### Indexar por fecha

```python
# indexar por fecha
data_HRES.index = pd.to_datetime(data_HRES['Date'])
data_HRES.head()
```

```python
# quitar columna Date
data_HRES.drop(columns=['Date'], inplace=True)
```

```python
paises_HRES = data_HRES.columns.to_list()
print(len(paises_HRES))
print(paises_HRES)
```

### Datos nulos

```python
# crear un datatframe del nÃºmero de valores nulos que tiene cada paÃ­s
nulos_HRES = pd.DataFrame(data_HRES.isnull().sum(), columns=['Num_Null'])

# parece que hay 2 casos, paÃ­ses con 199 nulos y paÃ­ses con todo (16861) nulos
nulos_HRES_pais1 = nulos_HRES[nulos_HRES['Num_Null'] == 199]
nulos_HRES_pais2 = nulos_HRES[nulos_HRES['Num_Null'] == 16861]
```

```python
# comprobamos que la suma total de las filas _pais1 y _pais2 es igual al de nulos_HRES
bool_HRES = nulos_HRES['Num_Null'].sum() == nulos_HRES_pais1['Num_Null'].sum() + nulos_HRES_pais2['Num_Null'].sum()
print(bool_HRES)

print("Paises con datos y algunos nulos")
print(nulos_HRES_pais1.index.tolist())
print()
print("Paises con todo nulos")
print(nulos_HRES_pais2.index.tolist())
```
