```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_HRIO = pd.read_csv('raw_data_Produccion_Hidraulica_Rios.csv',
                         skiprows=61)
data_HRIO.head()
```

## 1) PREPROCESAMIENTO DATOS

### Indexar por fecha

```python
# indexar por fecha
data_HRIO.index = pd.to_datetime(data_HRIO['Date'])
data_HRIO.head()
```

```python
# quitar columna Date
data_HRIO.drop(columns=['Date'], inplace=True)
```

```python
paises_HRIO = data_HRIO.columns.to_list()
print(f'{len(paises_HRIO)} paÃ­ses')
print(paises_HRIO)
```

### Datos nulos

```python
# crear un datatframe del nÃºmero de valores nulos que tiene cada paÃ­s
nulos_HRIO = pd.DataFrame(data_HRIO.isnull().sum(), columns=['Num_Null'])
nulos_HRIO['Num_Zero'] = nulos_HRIO.isin([0]).sum()
#nulos_HRIO

# HAY PAISES CON ALGUNOS NULOS (199) Y OTROS CON TODOS NULOS (16861)
```

```python
# parece que hay 2 casos, paÃ­ses con 199 nulos y paÃ­ses con todo (16861) nulos
nulos_HRIO_pais1 = nulos_HRIO[nulos_HRIO['Num_Null'] == 199]
nulos_HRIO_pais2 = nulos_HRIO[nulos_HRIO['Num_Null'] == 16861]

# comprobamos que la suma total de las filas _pais1 y _pais2 es igual al de nulos_HRIO
bool_HRIO = nulos_HRIO['Num_Null'].sum() == nulos_HRIO_pais1['Num_Null'].sum() + nulos_HRIO_pais2['Num_Null'].sum()
print(f'{bool_HRIO} coinciden las cifras')
print()
print(f"Paises los 199 primero dÃ­as nulos, el resto completo: {len(nulos_HRIO_pais1)}")
print(nulos_HRIO_pais1.index.tolist())
print()
print(f"Paises con todo nulos: {len(nulos_HRIO_pais2)}")
print(nulos_HRIO_pais2.index.tolist())
```
