```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_ETERR = pd.read_csv('Datos_Produccion_Energia_Eolica_Terrestre.csv',
                         skiprows=52)
data_ETERR.head()
```

## 1) PREPROCESAMIENTO DATOS

### Indexar por fecha

```python
# indexar por fecha
data_ETERR.index = pd.to_datetime(data_ETERR['Date'])
data_ETERR.head()
```

```python
# quitar columna Date
data_ETERR.drop(columns=['Date'], inplace=True)
```

```python
paises_ETERR = data_ETERR.columns.to_list()
print(len(paises_ETERR))
print(paises_ETERR)
```

### Datos nulos

```python
# crear un datatframe del nÃºmero de valores nulos que tiene cada paÃ­s
nulos_ETERR = pd.DataFrame(data_ETERR.isnull().sum(), columns=['Num_Null'])
#nulos_ETERR
```

```python
# parece que no hay datos nulos, pero veamos si hay ceros
nulos_ETERR['Num_Zero'] = data_ETERR.isin([0]).sum()
nulos_ETERR

# hay algunos paÃ­ses con todo ceros
nulos_ETERR_pais1 = nulos_ETERR[nulos_ETERR['Num_Zero'] == 0]
nulos_ETERR_pais2 = nulos_ETERR[nulos_ETERR['Num_Zero'] == 16861]
```

```python
# comprobamos que la suma total de las filas _pais1 y _pais2 es igual al de nulos_ETERR
bool_ETERR = nulos_ETERR['Num_Null'].sum() == nulos_ETERR_pais1['Num_Null'].sum() + nulos_ETERR_pais2['Num_Null'].sum()
print(bool_ETERR)

print("Paises sin nulos")
print(nulos_ETERR_pais1.index.tolist())
print()
print("Paises con todo nulos")
print(nulos_ETERR_pais2.index.tolist())
```
