```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_EMAR = pd.read_csv('Datos_Produccion_Energia_Eolica_Marina.csv',
                         skiprows=52)
data_EMAR.head()
```

## 1) PREPROCESAMIENTO DATOS

### Indexar por fecha

```python
# indexar por fecha
data_EMAR.index = pd.to_datetime(data_EMAR['Date'])
data_EMAR.head()
```

```python
# quitar columna Date
data_EMAR.drop(columns=['Date'], inplace=True)
```

### Países con costa

```python
paises_EMAR = data_EMAR.columns.to_list()
print(len(paises_EMAR))
print(paises_EMAR)
```

```python
# comparar los países entre estas dos listas
lista_1 = ['AL', 'AT', 'BA', 'BE', 'BG', 'CH', 'CY', 'CZ', 'DE', 'DK', 'EE', 'EL', 'ES', 'FI', 'FR', 'HR', 'HU', 'IE', 'IS', 'IT', 'LI', 'LT', 'LU', 'LV', 'ME', 'MK', 'MT', 'NL', 'NO', 'PL', 'PT', 'RO', 'RS', 'SE', 'SI', 'SK', 'TR', 'UK']
lista_2 = ['AL', 'BA', 'BE', 'BG', 'CY', 'DE', 'DK', 'EE', 'EL', 'ES', 'FI', 'FR', 'HR', 'IE', 'IT', 'LT', 'LV', 'ME', 'ML', 'NL', 'NO', 'PL', 'PT', 'RO', 'SE', 'SI', 'UK']

# veamos qué países de lista_2 no están en lista_1
paises_faltantes = []
for pais in lista_1:
    if pais not in lista_2:
        paises_faltantes.append(pais)
print(paises_faltantes)

paises_extra = []
for p in lista_2:
    if p not in lista_1:
        paises_extra.append(p)
print(paises_extra)
```

### Datos nulos

```python
# crear un datatframe del número de valores nulos que tiene cada país
nulos_EMAR = pd.DataFrame(data_EMAR.isnull().sum(), columns=['Num_Null'])
#nulos_EMAR
```

```python
# parece que no hay datos nulos, pero veamos si hay ceros
nulos_EMAR['Num_Zero'] = data_EMAR.isin([0]).sum()
nulos_EMAR
```

```python
# hay algunos países con todo ceros
nulos_EMAR_pais1 = nulos_EMAR[nulos_EMAR['Num_Zero'] == 0]
nulos_EMAR_pais2 = nulos_EMAR[nulos_EMAR['Num_Zero'] == 16861]
```

```python
# comprobamos que la suma total de las filas _pais1 y _pais2 es igual al de nulos_EMAR
bool_EMAR = nulos_EMAR['Num_Null'].sum() == nulos_EMAR_pais1['Num_Null'].sum() + nulos_EMAR_pais2['Num_Null'].sum()
print(bool_EMAR)

print("Paises sin nulos")
print(nulos_EMAR_pais1.index.tolist())
print()
print("Paises con todo nulos")
print(nulos_EMAR_pais2.index.tolist())
```
