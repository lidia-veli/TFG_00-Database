```python
import pandas as pd
import numpy as np
```

## Importar datos

```python
# DATOS PRODUCCIÓN EÓLICA TERRESTRE
df_PET = pd.read_csv('data_PET_1980_2024.csv', index_col=0, parse_dates=True)
                                    # tomamos la primer columna del csv como índice  en formato fecha
# DATOS PRODUCCIÓN EÓLICA MARINA
df_PEM = pd.read_csv('data_PEM_1980_2024.csv', index_col=0, parse_dates=True)


# Verificar si los índices (fechas) son iguales entre ambos DataFrames
fechas_coinciden = df_PET.index.equals(df_PEM.index)
if fechas_coinciden:
    print("Las fechas coinciden perfectamente entre los dos DataFrames.")
else:
    print("Las fechas no coinciden entre los dos DataFrames.")
```

```python
lista_paises_PET = df_PET.columns.tolist()
lista_paises_PEM = df_PEM.columns.tolist()

print(f'Paises con datos PET: {len(lista_paises_PET)}')
print(lista_paises_PET)
print()
print(f'Paises con datos PEM: {len(lista_paises_PEM)}')
print(lista_paises_PEM)
```

## Cobertura conjunta Europa

```python
PAISES_EU27 = [
    'AT', 'BE', 'BG', 'CY', 'CZ', 'DE', 'DK', 'EE', 'EL', 'ES', 'FI', 'FR', 
    'HR', 'HU', 'IE', 'IT', 'LT', 'LU', 'LV', 'MT', 'NL', 'PL', 'PT', 'RO', 
    'SE', 'SI', 'SK']

# tenemos que diferenciar países sin mar, que no tienen energía eólica marina
paises_EU27_sin_mar = ['AT', 'CZ', 'HU', 'LU', 'SK']
print(len(paises_EU27_sin_mar))

paises_EU27_mar = [
    'BE', 'BG', 'CY', 'DE', 'DK', 'EE', 'EL', 'ES', 'FI', 'FR', 'HR', 'IE', 
    'IT', 'LT', 'LV', 'MT', 'NL', 'PL', 'PT', 'RO', 'SE', 'SI']
print(len(paises_EU27_mar))
```

```python
df_prod_eolic_EU27 = pd.DataFrame(index=df_PET.index)

# para cada país
for pais in PAISES_EU27:
    if pais in paises_EU27_mar:  # si tiene costa
        # sumamos la producción eólica terrestre y marina
        df_prod_eolic_EU27[pais] = df_PET[pais] + df_PEM[pais]

    elif pais in paises_EU27_sin_mar:  # si no tiene costa
        # solo tomamos la producción eólica terrestre
        df_prod_eolic_EU27[pais] = df_PET[pais]
    # gestión de errores
    else:
        print(f"Error: El país {pais} no está en la lista de países de la UE27.")

# finalmente obtenemos la suma de todos los países de la UE27
df_prod_eolic_EU27['EU27'] = df_prod_eolic_EU27.sum(axis=1)

df_prod_eolic_EU27.tail()
```

```python
df_prod_eolic_EU27.to_csv('data_Prod_Eolic_TT_EU27_1980_2024.csv', index=True)
```
