```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_DE = pd.read_csv('raw_data_Demanda_Electrica.csv',
                         skiprows=53)
data_DE.head()
```

## 1) PREPROCESAMIENTO DATOS

### Indexar por fecha

```python
# indexar por fecha
data_DE.index = pd.to_datetime(data_DE['Date'])
data_DE.head()
```

```python
# quitar columna Date
data_DE.drop(columns=['Date'], inplace=True)
```

```python
# lista de paÃ­ses en los datos
paises_DE = data_DE.columns.to_list()
print(f'{len(paises_DE)} paises')
print(paises_DE)
```

### Datos nulos
en estos datos no hay nulos ni vacÃ­os

```python
# crear un datatframe del nÃºmero de valores nulos que tiene cada paÃ­s
nulos_DE = pd.DataFrame(data_DE.isnull().sum(), columns=['Num_Null'])
#nulos_DE

# parece que no hay datos nulos, pero veamos si hay ceros
nulos_DE['Num_Zero'] = nulos_DE.isin([0]).sum()
#nulos_DE

# NO HAY
```

## Exportar datos 1980 - 2024

```python
# queremos los datos del 1979-01-01 hasta el 2024-12-31
data_DE_csv = data_DE.loc['1980-01-01':'2024-12-31']
data_DE_csv.head()
```
