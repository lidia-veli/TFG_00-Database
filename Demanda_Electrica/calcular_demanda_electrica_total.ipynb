```python
import pandas as pd
import numpy as np
```

## Importar datos

```python
# DATOS CONSUMO ELÉCTRICO
df_DE = pd.read_csv('data_DE_1980_2024.csv', index_col=0, parse_dates=True)
                                    # tomamos la primer columna del csv como índice  en formato fecha
```

```python
paises_consumo = df_DE.columns.tolist()
PAISES_EU27 = [
    'AT', 'BE', 'BG', 'CY', 'CZ', 'DE', 'DK', 'EE', 'EL', 'ES', 'FI', 'FR', 
    'HR', 'HU', 'IE', 'IT', 'LT', 'LU', 'LV', 'MT', 'NL', 'PL', 'PT', 'RO', 
    'SE', 'SI', 'SK']

paises_EU27_faltan = []
for p in PAISES_EU27:
    if p not in paises_consumo:
        paises_EU27_faltan.append(p)
    else:
        continue
print(f'paises EU27 que no están en el dataset: {paises_EU27_faltan}')
```

## Gestionar Malta (``'MT'``)
que está en los datos de producción como uno de los países de la UE27, pero no en los de demanda.

```python
df_demanda_EU27 = df_DE.copy()

# crear la columna 'MT' y rellenarla con ceros
df_demanda_EU27['MT'] = 0

# ya que estamos, cambiamos todos los NaN del dataframe por 0.0s
df_demanda_EU27.fillna(0.0, inplace=True)

df_demanda_EU27.head()
```

## Cobertura conjunta Europa

```python
PAISES_EU27 = [
    'AT', 'BE', 'BG', 'CY', 'CZ', 'DE', 'DK', 'EE', 'EL', 'ES', 'FI', 'FR', 
    'HR', 'HU', 'IE', 'IT', 'LT', 'LU', 'LV', 'MT', 'NL', 'PL', 'PT', 'RO', 
    'SE', 'SI', 'SK']
```

```python
# vamos a dejar sólo los países en la lista de PAISES_UE27
for pais in df_demanda_EU27.columns:
    if pais not in PAISES_EU27:
        df_demanda_EU27.drop(pais, axis=1, inplace=True)
    else:
        continue

# y creamos la columna de la producción solar conjunta de los países de la UE27
df_demanda_EU27['EU27'] = df_demanda_EU27.sum(axis=1)

df_demanda_EU27.tail()
```

```python
df_demanda_EU27.to_csv('../data_Demanda_Electrica_EU27_1980_2024.csv', index=True)
```
